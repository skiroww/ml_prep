# Part 2 - Segmentation.

## I'll start with KMeans clustering. 

### Pros of using it:


```python
'''
Let's now use sklearn scaling, PCA in order to 
quickly and conviniently apply kMeans 

Firstly - find all continuous features to scale.

UPD - It was found to provide inferior results to unscaled
data -> I did not use it and only used
'''
float_cols = [x for x in df.columns if df[x].dtype==np.float64]
ss = StandardScaler()
for i in float_cols:
    df[i] = ss.fit_transform(np.array(df[i]).reshape(-1, 1))
```

```python
'''
Now, let us use PCA for solving such issues as high  dimensionality and mutual 
correlation. I will use the number of components saving 0.9 data variance - both reducing
the dimensiontionality and saving decent amount of variance in data.
'''
pc = PCA(n_components = 0.9,svd_solver='full')
new_x = pc.fit_transform(df)
len(new_x[0])
```

```python
type(new_x)
```

```python
'''
Use elbow method to determine the best number of clusters.

'''


wcss = []
import gc 
# I would use a garbage collector delete everything here 
gc.collect()
# fit the k-means model for different k values and calculate the WCSS
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='random',random_state=228,n_jobs=1)
    kmeans.fit(new_x)
    wcss.append(kmeans.inertia_)
    gc.collect()

# plot the WCSS values against the number of clusters
_ = plt.figure(figsize = (15,5), dpi=70)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

```

```python
'''
Now, it's obvious that we need 3 to 5 clusters for best performance.
Let's use 3 in order to get the best explainable data.
After that, use PCA and t-sne with 2 components for visualization
purposes.

'''
```

```python
'''
Now, visualize and create a new columns for post-analysis.

'''


kmeans = KMeans(n_clusters=3, init='k-means++',n_jobs=2)
kmeans.fit(new_x)
plt.figure(figsize=(8, 6))
new_x = PCA(n_components = 2,svd_solver='full').fit_transform(new_x)
sns.scatterplot(x=new_x[:, 0], y=new_x[:, 1],hue=kmeans.labels_)
plt.title('K-means clustering with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
df['Clusters'] = kmeans.labels_
```

```python
df = pd.read_csv('FINAL DATA.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)

df['Clusters'] = kmeans.labels_
cl1 = df[df.Clusters ==0]
cl2 = df[df.Clusters ==1]
cl3 = df[df.Clusters ==2]
```
