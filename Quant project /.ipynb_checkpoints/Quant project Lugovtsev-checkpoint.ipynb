# 1 - The mean, median and average estimation

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
 import pandas as pd
import numpy as np

def estimate_next_day_volume(data, k):
    
    # Create a new dataframe with the rolling mean and median of the trading volume over the last k days
    rolling_mean = data.rolling(window=k).mean().dropna()
    rolling_median = data.rolling(window=k).median().dropna()
    
    # Compute the estimated volume for each day using both the mean and median
    mean_estimates = data.shift(1).loc[rolling_mean.index] * rolling_mean / rolling_mean.shift(1)
    median_estimates = data.shift(1).loc[rolling_median.index] * rolling_median / rolling_median.shift(1)
    
    # Compute the relative error for each estimation and take the absolute value
    mean_error = np.abs(mean_estimates - data.loc[mean_estimates.index]) / data.loc[mean_estimates.index]
    median_error = np.abs(median_estimates - data.loc[median_estimates.index]) / data.loc[median_estimates.index]
    
    # Compute the average accuracy of both estimations over the whole 5-year period
    accuracy_mean = (1 - mean_error.mean()) 
    accuracy_median = (1 - median_error.mean())
    average_accuracy = (accuracy_mean + accuracy_median) / 2
    
    return np.array([average_accuracy,accuracy_mean,accuracy_median])
```

```python
def estimate_next_day_volume_and_accuracy(volume_data, k):
    total_error_mean = 0
    total_error_median = 0
    for i in range(k, len(volume_data)):
        last_k_days_volumes = volume_data[i-k:i]
        avg_volume = sum(last_k_days_volumes) / k
        median_volume = sorted(last_k_days_volumes)[k // 2]
        estimated_volume_avg = (avg_volume + median_volume) / 2
        error_mean = abs(avg_volume - volume_data[i]) / volume_data[i]
        error_median = abs(median_volume - volume_data[i]) / volume_data[i]
        total_error_mean += error_mean
        total_error_median += error_median
    average_accuracy_mean = (1 - total_error_mean / (len(volume_data) - k)) 
    average_accuracy_median= (1 - total_error_median / (len(volume_data) - k)) 
    return [(average_accuracy_mean+average_accuracy_median)/2,average_accuracy_mean,average_accuracy_median]
```

```python
def k_days_loop(data):
    acc_mean = []
    acc_median = []
    acc_average = []
    for i in range (1,1001):
        acc = estimate_next_day_volume_and_accuracy(data,i)
        acc_average.append(acc[0])
        acc_mean.append(acc[1])
        acc_median.append(acc[2])
    return [acc_average,acc_mean,acc_median]
```

# 2 - The weighted average estimation

```python
def predict_next_day_volume(volume_data, k):
    """
    Predicts the next day's trading volume based on the weighted average of the last k days of volume data.
    
    Args:
        volume_data (list): A list of trading volume data for a period of time.
        k (int): The number of days to use in the weighted average.
        
    Returns:
        float: The predicted volume for the next day.
    """
    if k > len(volume_data):
        raise ValueError("k must be less than or equal to the length of the volume data")
    
    # Calculate the weighted average
    weights = range(1, k+1)
    weighted_sum = sum([w * v for w, v in zip(weights, volume_data[-k:])])
    weighted_avg = weighted_sum / sum(weights)
    
    return weighted_avg


def average_accuracy(volume_data, k):
    """
    Calculates the average accuracy of the prediction of the next day's trading volume based on the weighted average of the 
    last k days of volume data.
    
    Args:
        volume_data (list): A list of trading volume data for a period of time.
        k (int): The number of days to use in the weighted average.
        
    Returns:
        float: The average accuracy of the predictions over the whole period of data.
    """
    if k > len(volume_data):
        raise ValueError("k must be less than or equal to the length of the volume data")
    
    # Calculate the predictions and accuracy for each day
    total_accuracy = 0.0
    for i in range(k, len(volume_data)):
        actual_volume = volume_data[i]
        predicted_volume = predict_next_day_volume(volume_data[:i], k)
        accuracy = abs(actual_volume - predicted_volume) / actual_volume
        total_accuracy += accuracy
    
    # Calculate the average accuracy
    average_accuracy = total_accuracy / (len(volume_data) - k)
    
    return average_accuracy
```

```python
def average_loop(data):
    acc = []
    for i in range(1,1001):
        acc.append(average_accuracy(data,i))
    return acc
```

```python
accuracy_weighted_gap = average_loop(gps)
accuracy_weighted_apple = average_loop(aapl)
accuracy_weighted_exxon = average_loop(xom)
```

# 3 - The linear regression estimation

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

def predict_next_day_volume_regr(data, k):
    # Initialize variables to keep track of predictions and accuracies
    predictions = []
    accuracies = []
    total_accuracy=0
    # Loop over each day in the data series
    for i in range(len(data)):
        # Check if there are at least k previous days of data available
        if i >= k:
            # Split the data into training and testing sets
            train_data = data.iloc[i-k:i]
            test_data = data.iloc[i:i+1]
            
            # Initialize the linear regression model
            model = LinearRegression()
            
            # Fit the model on the last k days of training data
            X_train = train_data.tail(k).index.to_numpy().reshape(-1, 1)
            y_train = train_data.tail(k).to_numpy().reshape(-1, 1)
            model.fit(X_train, y_train)
            
            # Make a prediction on the next day's volume using the model
            X_test = test_data.index.to_numpy().reshape(-1, 1)
            y_pred = model.predict(X_test)[0][0]
            predictions.append(y_pred)
            
            # Compute the accuracy of the prediction using mean absolute error
            y_true = test_data.to_numpy()[0]
            accuracy = abs(y_true - y_pred) / y_true
            total_accuracy += accuracy
           
            
    
    # Compute the average accuracy over the 5 year period
    avg_accuracy = total_accuracy / (len(data) - k)
    
    return avg_accuracy
```

```python
def regr_loop(data):
    acc = []
    for i in range(1,1001):
        acc.append(predict_next_day_volume_regr(data,i))
    return acc
```

```python
accuracy_regr_gap = regr_loop(gps)
accuracy_regr_apple = regr_loop(aapl)
accuracy_regr_exxon = regr_loop(xom)
```

## Small adjustment - use alpha * mean + beta * median combination for estimating the next day's volume where alpha and beta are coefficients ranging from 0.1 to 0.9

```python
from statistics import median, mean

def estimate_next_day_volume(volumes, k):
    total_accuracy = 0.0
    
    for coeff in range(1, 10):
        coeff /= 10.0
        for i in range(k, len(volumes)-1):
            last_k_volumes = volumes[i-k:i]
            median_volume = median(last_k_volumes)
            mean_volume = mean(last_k_volumes)
            estimated_volume = coeff*mean_volume + (1-coeff)*median_volume
            actual_volume = volumes[i+1]
            accuracy = 1 - abs(estimated_volume - actual_volume) / actual_volume
            total_accuracy += accuracy
    average_accuracy = total_accuracy / ((len(volumes)-k-1) * 9)
    return average_accuracy


def coeff_loop(data):
    acc = []
    for i in range(1,1001):
        acc.append(estimate_next_day_volume(data,i))
    return acc
```
