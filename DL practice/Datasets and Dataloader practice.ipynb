```python
from torch.utils.data import Dataset
from torchvision import datasets
from matplotlib import pyplot as plt
import torch 
from torchvision.transforms import ToTensor
```

```python
training_data = datasets.FashionMNIST(root = 'Documents/GitHub/ml_prep/DL practice', train = True
                                     , download = True, transform = ToTensor())
```

```python
labels_map = {
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}
figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(high = len(training_data), size=(1,)).item()
    # the randint returns a tensor of random integers, size is crucial
    img, label = training_data[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(labels_map[label])
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="gray")
plt.show()
```
